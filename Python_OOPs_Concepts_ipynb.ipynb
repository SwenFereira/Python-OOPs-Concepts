# **Python OOPs Concepts:**


There are 2 types of programming approch,

1) **Procedural Programming** is an approch where we use functions and our own logic to build the code.

2) **Object Oriented Programming** where the basic idea is to use classes and objects to represent real world concepts and entities.

image.png

A **class** is blueprint or template for creating objects and defines the properties and methods that an object of that class will have.

An **object** is an instance of a class which has its own data and methods.

Properties are data or state of an object and **Methods** are the actions or behavior that an object can perform.

**Encapsulation** - In oop, Encapsulation is a key feature which means the internal state of object is hidden and can only be accessed or modified through the objects methods. This helps to protect the objects data's and prevent it from being modified in unexpected ways.
